A futures contract is a standardized agreement that is traded on an exchange. It involves a seller who commits to delivering a specific quantity of the underlying asset at a predetermined price on a future date. When it comes to cotton futures, the contract represents 50,000 pounds of cotton.

Cotton futures are actively traded on NYMEX, which is part of the CME Group. On this platform, the contract is settled financially. Additionally, cotton futures can also be traded on ICE, where the settlement is achieved by physically delivering the cotton.

Factors that affect the prices of cotton include: <br>

~ Reduced output from United States, India, and China <br>
~ The price of substitute fabrics <br>
~ The demand for cotton-based products <br>
~ Energy prices <br>
~ Chinaâ€™s stockpiling policy (Subsidies can lower the price of the commodity, while a ban on importation can increase prices.)

## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import yfinance as yf
import warnings
warnings.filterwarnings('ignore')
```

## Data Preparation

```python
# Download historical price data for cotton futures ('CT=F')
data = yf.download('CT=F')[['Close','Open']]

# Calculate the simple moving average (SMA) over a 200-day window
data['sma'] = data['Close'].rolling(window=200).mean()

# Define the period for calculating high and low values
n = 20
data['high1'] = 0
data['low1'] = 0

for i in range(n, len(data)):
	period = data['Close'][i-n:i]
	data['high1'][i] = max(period)
	data['low1'][i] = min(period)

# Calculate the rolling high and low values using vectorized operations
data['high'] = data['Close'].rolling(window=n).max()
data['low'] = data['Close'].rolling(window=n).min()
```

## Strategy 

```python
data['regime'] = 0  
z = 0  
pop = 0  

# try it with data[high], you'll get a different result
for i in range(0, len(data)-1):
    if data['Close'][i] >= data['high1'][i] and data['Close'][i] <= data['sma'][i] and z == 0:
        data['regime'][i+1] = 1  
        z = 1  # Set 'z' to 1 to indicate that a regime has started
        pop = i  # Set 'pop' to the current time index 'i' index of a previous regime
    elif i > (pop+5) and z == 1:
        data['regime'][i+1] = -1  
        z = 0  # Set 'z' back to 0 to indicate the end of the regime
        pop = 0  # Reset 'pop' to zero
```

## Entry and exit prices

```python
buy_price = []
sell_price = []
start = []
end = []
open = 0

for i in range(0, len(data)):
	if data['regime'][i] == 1:
		buy_price.append(data['Open'][i])
		open = data['Open'][i]
		start.append(i)
	elif data['regime'][i] == -1:
		sell_price.append(data['Open'][i])
		end.append(i)

if len(buy_price) != len(sell_price):
	sell_price.append(data['Close'][-1])
	end.append(len(data))
```

## Return calculation and P&L plot

```python
results = pd.DataFrame()
results['Open'] = buy_price
results['Close'] = sell_price
results['Return'] = ((results['Close'] / results['Open'] - 1) * -1) + 1
returns = results['Return'].cumprod()

results['start'] = start
results['end'] = end
results['len'] = results['end'] - results['start']

plt.figure(figsize=[10,8])
plt.plot(returns, label='Strategy')
plt.title('Cotton Short Trading Strategy Equity Curve')
plt.legend()
plt.show()
print("Return:", round(returns.iloc[-1]/returns.iloc[0] -1,4)*100,"%")
```
